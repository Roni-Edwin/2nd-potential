```julia
using IntervalArithmetic

function check(func,interval_check)
    lower=interval_check[1]
    upper=interval_check[2]
    beg=lower
    print(beg, "\n")
    last=upper+1
    temp=beg..last
    while beg<upper+1/(2^(30))
        res=func(temp)
        iters=0
        lim=0..0
        while !(((res ∩ lim)==∅) & (res>lim))
            last=(beg+last)/2
            temp=beg..last
            res=func(temp)
            iters=iters+1
            if iters>=20
                last=nextfloat(beg)
                #print(beg," ",last, "\n")
                iters=0
            end
        end
        print("Checked from ",lower, " up to ",beg, "\n")  
        beg=last
        last=beg+1
        temp=beg..last
    end
    print("Checked on interval ", "[",lower,",",upper,"]","\n")
end
    

    
```

```julia
alphas=[6,8,10,12] #1.408
lp_salphas=[7.83668,11.813148, 15.830697, 19.854857]
up_salphas=[7.83669, 11.813151, 15.830699, 19.854859]


for j in [1,2,3,4]
    print('\n','\n','\n')
    alpha=alphas[j]
    s_alphar=lp_salphas[j]..up_salphas[j]
    function F_a(x)
        1/(1+s_alphar*(x^alpha))
    end
    
    function dF_a(x)
        if x==0
            val=0
        else
            val=-(alpha*F_a(x)*(1-F_a(x)))/x
        end
    end
    
    function A_a(x)
        fac1=(4..4)/((alpha-5)*(x^2))
        fac2=(alpha-(4..4))/3+(alpha-(5..5)+(3*x)/2)/(5*((3*x/2)^(alpha-4)))
        res=fac1*fac2
        fac3=(alpha-(5..5)+x/2)/(3*x*((x/2)^(alpha-4))*(alpha-5))
        res+4*fac3
    end
    
    function C_a(x)
        fac1=(16..16)/((alpha-5)*(x^2))
        fac2=(alpha-(4..4))/9+(alpha-(5..5)+(3*x)/2)/(25*((3*x/2)^(alpha-4)))
        res=fac1*fac2
        fac3=(alpha-(5..5)+x/2)/(9*((x/2)^(alpha-4))*(alpha-5))
        res+16*fac3
    end
    
    
    value=2*(alpha+3)*A_a(8)/(s_alphar)+4*C_a(8)/s_alphar+192*alpha*F_a(7.5)
    for n=1:7
        value=value+2*(3*n^2*F_a(n)+n^3*dF_a(n))
    end
    print("Value is ", value, '\n','\n')
    
    function H(x,n)
        F_a(n)/((x-n)^2)+dF_a(n)/(x-n)
    end
    
    p=30
    function func_tobound(x)
        val=((4..4)*alpha/3+8)*((1/(p+1))^alpha)
        val=-val*(alpha+p)/(alpha-1)-H(x,0)
        #val=val=F_a(p)*(1/(p-x)+4*alpha*log(6*x)/x)
        #print("val is ",val,'\n')
        for n=1:p
            if n==round(x)
                val=val-H(x,-n)
            else
                val=val-H(x,n)-H(x,-n)
            end
        end
        val
    end
    
    
    print("On alpha = ",alpha, '\n')
    #print(func_tobound(1.5),'\n')
    check(func_tobound,[1.5,8])
end
```
