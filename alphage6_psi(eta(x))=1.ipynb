```julia
using IntervalArithmetic

function check(func,interval_check)
    lower=interval_check[1]
    upper=interval_check[2]
    beg=lower
    print(beg, "\n")
    last=upper+1
    temp=beg..last
    while beg<upper+1/(2^(30))
        res=func(temp)
        iters=0
        lim=0..0
        while !(((res ∩ lim)==∅) & (res>lim))
            last=(beg+last)/2
            temp=beg..last
            res=func(temp)
            iters=iters+1
            if iters>=20
                last=nextfloat(beg)
                #print(beg," ",last, "\n")
                iters=0
            end
        end
        print("Checked from ",lower, " up to ",beg, "\n")  
        beg=last
        last=beg+1
        temp=beg..last
    end
    print("Checked on interval ", "[",lower,",",upper,"]","\n")
end
    

    
```

```julia
alphas=[4,6,8,10]
lp_salphas=[4,7.83668,11.813148, 15.830697, 19.854857]
up_salphas=[4,7.83669, 11.813151, 15.830699, 19.854859]


for j in [2,3,4]
    alpha=alphas[j]
    s_alphar=lp_salphas[j]..up_salphas[j]
    function F_a(x)
        1/(1+s_alphar*(x^alpha))
    end
    
    function dF_a(x)
        -(alpha*F_a(x)*(1-F_a(x)))/x
    end
    
    p=50
    numb=2*(F_a(1/2)-1)+((pi..pi)^2*F_a(1/2))/6
    fac1=((4..4)*(alpha+1))/(3*(p+1)^alpha)
    fac2=((alpha..alpha)+p)/((alpha..alpha)-1)
    numb=numb-fac1*fac2
    #print(Talpha, '\n', '\n')
    for n=1:p
        termn=1/(n^2)+(alpha*(1-F_a(n)))/(n^2-1/4)
        numb=numb -F_a(n)*termn
    end
    print("value of number on alpha = ",alpha, " is ",numb, '\n','\n','\n')
end
```

```julia
alphas=[4,6,8,10,12]
lp_salphas=[4,7.83668,11.813148, 15.830697, 19.854857]
up_salphas=[4,7.83669, 11.813151, 15.830699, 19.854859]

function scaling_a(alpha)
    if alpha<=12
        for j in [2,3,4,5]
            if alphas[j]==alpha
                scale_a=lp_salphas[j]..up_salphas[j]
            end
        end
    else
        bb=alpha-2
        accb=bb+sqrt((bb..bb)^2-3)
        err=((21..21)/10)*(alpha+1)^3
        err=err/(alpha-1)
        err=err*(1/2)^(alpha-1)
        lo=accb-err
        up=accb+err
        scale_a=lo..up
    end
    scale_a
end

```
