```julia
using IntervalArithmetic

function check(func,interval_check)
    lower=interval_check[1]
    upper=interval_check[2]
    beg=lower
    print(beg, "\n")
    last=upper+1
    temp=beg..last
    while beg<upper+1/(2^(30))
        res=func(temp)
        iters=0
        lim=0..0
        while !(((res ∩ lim)==∅) & (res>lim))
            last=(beg+last)/2
            temp=beg..last
            res=func(temp)
            iters=iters+1
            if iters>=20
                last=nextfloat(beg)
                #print(beg," ",last, "\n")
                iters=0
            end
        end
        print("Checked from ",lower, " up to ",beg, "\n")  
        beg=last
        last=beg+1
        temp=beg..last
    end
    print("Checked on interval ", "[",lower,",",upper,"]","\n")
end
    

    
```

```julia
using IntervalArithmetic


function F4(x)
    1/(4*(x..x)^4+1)
end

function dF4(x)
    -16*(x..x)^3/((4*(x..x)^4+1)^2)
end

function d2F4(x)
    4*F4(x)*((4*x^2)/(4*x^4+1))*(4*(1-2*F4(x))+1)
end

function H4(x,n)
    if abs(x-n)<=1/32
        if n==0
            val=d2F4(x)*0.5
        else
            u=0.5(x+n+abs(x-n))
            l=x+n-u
            val=4*F4(u)*((1-F4(l))/(u^2))*(4*(1-2*F4(l))+1)
            val=0.5*val
        end
    else
        val=(F4(x)-F4(n)-dF4(n)*(x-n))/((x-n)^2)
    end
    val
end


p=3000

function mustbepos(x)
    val=H4(x,0)
    for n=1:p
        val=val+H4(x,-n)+H4(x,n)
    end
    val-2*(F4(x)+5*F4(p))/(p-(x..x))
end
    
#Have checked up to 29.5
check(mustbepos,[0,9],0)
        
    
```
