```julia
using IntervalArithmetic

function check(func,interval_check)
    lower=interval_check[1]
    upper=interval_check[2]
    beg=lower
    print(beg, "\n")
    last=upper+1
    temp=beg..last
    while beg<upper+1/(2^(30))
        res=func(temp)
        iters=0
        lim=0..0
        while !(((res ∩ lim)==∅) & (res>lim))
            last=(beg+last)/2
            temp=beg..last
            res=func(temp)
            iters=iters+1
            if iters>=20
                last=nextfloat(beg)
                #print(beg," ",last, "\n")
                iters=0
            end
        end
        print("Checked from ",lower, " up to ",beg, "\n")  
        beg=last
        last=beg+1
        temp=beg..last
    end
    print("Checked on interval ", "[",lower,",",upper,"]","\n")
end
    

    
```
